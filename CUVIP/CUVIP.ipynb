```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = pd.read_csv('cuvip.csv')
```

```python
file.head()
```

```python
grouped_data = file.groupby(['year', 'name'])
display(grouped_data.head())
```

```python
data_by_year = {}
for year, data in grouped_data:
    if year[0] not in data_by_year:
        data_by_year[year[0]] = []
    data_by_year[year[0]].append(data)
print(data_by_year[1959])
```

```python
top_gold_medal = {}
for year, data in data_by_year.items():
  for i in data:
    if i['name'].values[0] not in top_gold_medal:
      top_gold_medal[i['name'].values[0]] = 0
    else: top_gold_medal[i['name'].values[0]] += i['gold'].values[0]
```

```python
colors = [
    '#1f77b4',  # Muted Blue
    '#ff7f0e',  # Safety Orange
    '#2ca02c',  # Cooked Asparagus Green
    '#d62728',  # Brick Red
    '#9467bd',  # Muted Purple
    '#8c564b',  # Chestnut Brown
    '#e377c2',  # Raspberry Pink
    '#7f7f7f',  # Middle Gray
    '#bcbd22',  # Curry Yellow-Green
    '#17becf'   # Blue-Teal
]
plt.bar(top_gold_medal.keys(), top_gold_medal.values(),color=colors)
plt.title("Top Gold Medalist")
plt.xticks(rotation=45)
plt.show()
```
