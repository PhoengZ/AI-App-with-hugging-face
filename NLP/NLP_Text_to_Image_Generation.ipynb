# Text-to-Image Generation
มีหลาย Model เช่น
Stable Diffusion Process
ซึ่งจะใช้กระบวนการสร้างภาพคืนจากสิ่งที่เบลอโดยการลบ noise เพื่อให้ตรงกับภาพที่ต้องการได้

```python
# model Stable diffusionm จะไม่มีอยู่ใน lib transformer แต่อยู่ใน StableDiffusionPipeline
from diffusers import StableDiffusionPipeline
import torch # ช่วยให้ run model ขนาดใหญ่ได้ดีขึันและทำ pararell ได้
from IPython.display import display # ใช้แสดงภาพ
```

```python
model = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model, torch_dtype=torch.float16)
# argument float16 ทำให้ program ใช้ ram น้อยลงเพราะใช้ความละเอียดจากเดิม 32 bit เป็น 16 bit
pipe = pipe.to("cuda") # ensure ว่ามันจะประมวลผลใช้ gpu
```

```python
prompt = "A cartoon man driving a flying car, japan digital art"
image = pipe(prompt).images[0]
display(image)
```
