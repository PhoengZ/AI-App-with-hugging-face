```python
from transformers import pipeline
import pandas as pd
```

```python
from google.colab import files
```

```python
uploaded = files.upload()
file_name = list(uploaded.keys())[0] # open upload files
df = pd.read_excel(file_name)
```

```python
df = df.astype(str) # Ensure all cell are string
```

```python
display(df.head()) # use display method
```

```python
# Initiate Table Answering
tableQA_model = "google/tapas-base-finetuned-wtq"
tableQA = pipeline('table-question-answering',model=tableQA_model)
```

```python
# Define the question
question = "What is district for 3905884"
```

```python
# Do the table qa for and show the answer
answer = tableQA(table=df, query=question) # using table to specific table for answering and
# using query for specific question
```

```python
print(f"Q: {question}")
print(f"A: {answer[0]['answer']}")
```
